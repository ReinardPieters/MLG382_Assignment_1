```python
def uniV(df):
    import pandas as pd
    students_df = pd.read_csv(r"C:\Student_performance_data.csv", index_col="StudentID")
    
    output_students_df = pd.DataFrame(columns=['Count', 'Missing', 'Unique', 'Dtype', 'Mode', 
                                     'Mean', 'Min', '25%', 'Median', '75%', 'Max', 'Std', 'Skew', 'Kurt'])
    
    for col in students_df:
        try:
            output_students_df.loc[col] = [
                students_df[col].count(), students_df[col].isnull().sum(), students_df[col].nunique(), students_df[col].dtype,
                students_df[col].mode().values[0], students_df[col].mean(), students_df[col].min(), students_df[col].quantile(0.25),students_df[col].median()
                , students_df[col].quantile(0.75),  students_df[col].max(),students_df[col].std(),students_df[col].skew(),students_df[col].kurt()
            ]
        except Exception as e:
            print(f"Error processing column {col}: {str(e)}")
            continue
    
    return output_students_df
    
    
```

```python
uniV(students_df)
```

```python
X = students_df.iloc[:, :-1]
y = students_df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

results = []

processed_features = set()

for feature in X_train.columns:
    if feature in processed_features:
        continue
        
    # Train model
    rf = RandomForestRegressor(random_state=123)
    rf.fit(X_train[[feature]], y_train)
    
    # Predict and calculate MSE
    y_pred = rf.predict(X_test[[feature]])
    mse = mean_squared_error(y_test, y_pred)
        
    results.append({
        'Feature': feature,
        'MSE': mse,
        'Feature_Values': X_train[feature].values[:10]  # Store sample values for verification
    })
    processed_features.add(feature)

feature_importance = pd.DataFrame(results).sort_values('MSE')
print(f"\nUnique features processed: {len(results)}/{len(X_train.columns)}")
```

```python
print("\nTop 5 most important features:")
feature_importance.head()
```

```python
print("\nBottom 5 least important features:")
feature_importance.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_importance = feature_importance.sort_values('MSE')

plt.figure(figsize=(12, 8))

sns.barplot(x='MSE', y='Feature', data=feature_importance, color='skyblue')

plt.title('Feature Importance by MSE Score', fontsize=16, pad=20)
plt.xlabel('Mean Squared Error (MSE)', fontsize=12)
plt.ylabel('Features', fontsize=12)

for i, (mse, feature) in enumerate(zip(feature_importance['MSE'], feature_importance['Feature'])):
    plt.text(mse, i, f' {mse:.3f}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
def biV(df, label):
    from scipy import stats
    import pandas as pd
    import numpy as np
    
    output_df = pd.DataFrame(columns=['Correlation', 'p-value'])
    
    for col in df:
        if col != label:
            if df[col].isnull().sum() == 0:
                if pd.api.types.is_numeric_dtype(df[col]):
                    r, p = stats.pearsonr(df[label], df[col])
                    output_df.loc[col] = [round(r, 3), round(p, 6)]
                else:
                    output_df.loc[col] = [np.nan, 'nulls']
    
    output_df.sort_values(by='Correlation', key=abs, ascending=False, inplace=True)
    return output_df

import pandas as pd
pd.options.display.float_format = '{:.5f}'.format
students_df = pd.read_csv("C:\\Student_performance_data.csv", index_col="StudentID")
result = biV(students_df, 'GradeClass')  # Fix: added closing parenthesis
result
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix
corr_matrix = students_df.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```
